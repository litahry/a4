```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation
import seaborn as sns
import matplotlib.pyplot as plt
from pyspark.sql.types import DoubleType, IntegerType
from pyspark.sql.functions import col

# Get the names of all numeric columns in the DataFrame
numeric_columns = [column for column, data_type in df.dtypes if data_type in ['double', 'int']]

# Convert IntegerType columns to DoubleType for compatibility with Correlation
for column in numeric_columns:
    if df.schema[column].dataType == IntegerType():
        df = df.withColumn(column, col(column).cast(DoubleType()))

# Assemble the selected numeric columns into a vector, handling null values by skipping
assembler = VectorAssembler(inputCols=numeric_columns, outputCol="features", handleInvalid="skip")
df_assembled = assembler.transform(df.select(numeric_columns))

# Calculate correlation matrix
corr_matrix = Correlation.corr(df_assembled, column="features", method='pearson').head()

# Extract correlation values
corr_values = corr_matrix[0].toArray()

# Plot correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_values, annot=True, cmap='coolwarm', xticklabels=numeric_columns, yticklabels=numeric_columns)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np

# Identify all columns of integer and double types
numeric_columns = [column for column, dtype in df.dtypes if dtype in ['int', 'double']]

# Iterate over each numeric column to calculate and plot histograms
for column in numeric_columns:
    # Convert the DataFrame column to an RDD
    column_rdd = df.select(column).rdd.flatMap(lambda x: x)

    # Handle nulls in RDD if there are any
    column_rdd = column_rdd.filter(lambda x: x is not None)

    # Get histogram data
    histogram_data = column_rdd.histogram(10) 

    # Extract bin edges and counts from the histogram data
    bin_edges = histogram_data[0]
    counts = histogram_data[1]

    # Calculate bin centers
    bin_centers = [0.5 * (bin_edges[i] + bin_edges[i+1]) for i in range(len(bin_edges)-1)]

    # Normalize counts to form a probability density
    total_count = float(sum(counts))
    density = [count / total_count for count in counts]

    # Plotting histogram for the current column
    plt.figure(figsize=(10, 6))
    plt.bar(bin_centers, density, width=np.diff(bin_edges), edgecolor='black', align='center')
    plt.plot(bin_centers, density, '-o')  # Connect the bars with a line

    plt.xlabel(column)
    plt.ylabel('Density')
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
from pyspark.sql.functions import col
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation
from pyspark.ml.regression import LinearRegression
from pyspark.ml.linalg import Vectors


# Mental health conditons
conditions = ["Insomnia", "Anxiety", "OCD", "Depression"]

for condition in conditions:
    # Assemble features into a vector
    assembler = VectorAssembler(inputCols=["Hours per day", condition], outputCol="features")
    df_filtered = df.na.drop(subset=["Hours per day", condition])  # Drop NA values
    assembled_df = assembler.transform(df_filtered)

    # Calculate correlation
    corr_matrix = Correlation.corr(assembled_df, "features").head()[0]
    correlation = corr_matrix[0, 1]  # Correlation coefficient

    # Fit a linear regression model
    lr = LinearRegression(featuresCol="features", labelCol=condition)
    lr_model = lr.fit(assembled_df)

    # Summary of the model
    summary = lr_model.summary
    print(f"{condition} vs Hours per Day:")
    print(f"Correlation coefficient: {correlation:.2f}")
    print(f"Slope: {lr_model.coefficients[1]}")
    print(f"Intercept: {lr_model.intercept}")
    print(f"P-value: {summary.pValues[1]:.5f}")
    print(f"Standard error: {summary.coefficientStandardErrors[1]:.5f}")

    # Prepare data for plotting
    hours = df_filtered.select("Hours per day").rdd.flatMap(lambda x: x).collect()
    values = df_filtered.select(condition).rdd.flatMap(lambda x: x).collect()

    # Generating prediction points for regression line
    min_hours, max_hours = min(hours), max(hours)
    prediction_df = spark.createDataFrame([
        (Vectors.dense([1, min_hours]),),
        (Vectors.dense([1, max_hours]),)
    ], ["features"])

    predictions = lr_model.transform(prediction_df).select("prediction").rdd.flatMap(lambda x: x).collect()

    # Plot
    plt.figure(figsize=(6, 4))
    plt.scatter(hours, values, alpha=0.5)
    plt.plot([min_hours, max_hours], predictions, color='red')  # Regression line
    plt.title(f"{condition} vs Hours per Day\nCorrelation: {correlation:.2f}, p-value: {summary.pValues[1]:.2e}")
    plt.xlabel('Hours per Day')
    plt.ylabel(condition)
    plt.show()
```

```python
# Correlation heatmap for string data using cramers v 
    
from pyspark.sql.functions import col, count, lit
from pyspark.sql import DataFrame
import numpy as np
from scipy.stats import chi2_contingency

def cramers_v(df: DataFrame, col1: str, col2: str):
    # Create a contingency table in PySpark
    df_crosstab = df.crosstab(col1, col2)
    
    # Convert to Pandas for chi-squared test
    contingency_table = df_crosstab.toPandas().set_index(df_crosstab.columns[0])

    # Ensure contingency table's data types are integers
    for column in contingency_table.columns:
        contingency_table[column] = contingency_table[column].astype(int)
    
    # Perform Chi-squared test
    chi2, _, _, _ = chi2_contingency(contingency_table)
    n = contingency_table.values.sum()  # Total number of observations

    # Calculation of Cramer's V
    phi2 = chi2 / n
    r, k = contingency_table.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    
    # Correction for degrees of freedom
    denom = min((kcorr-1), (rcorr-1))
    if denom <= 0:
        return 0
    else:
        return np.sqrt(phi2corr / denom)

# Categorical columns
categorical_columns = [col for col, dtype in df.dtypes if dtype == 'string']

# Create an empty dictionary to store correlation values
corr_matrix = {}

# Calculate Cramer's V for every pair of categorical variables
for col1 in categorical_columns:
    corr_matrix[col1] = {}
    for col2 in categorical_columns:
        if col1 != col2:
            corr_value = cramers_v(df, col1, col2)
            corr_matrix[col1][col2] = corr_value
        else:
            corr_matrix[col1][col2] = 1.0  # Self-correlation is always 1

# Convert dictionary to Pandas DataFrame for easier plotting
corr_matrix_df = pd.DataFrame(corr_matrix)

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix_df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Cramer\'s V Correlation Matrix')
plt.show()
```

```python
from pyspark.sql.functions import col, when, count, lit, isnan, isnull, trim, length
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Identify columns with missing values and calculate the percentage of missing values
def missing_values_count(df):
    numeric_cols = [c for c, dtype in df.dtypes if dtype in ('int', 'double')]
    non_numeric_cols = [c for c, dtype in df.dtypes if dtype not in ('int', 'double')]

    numeric_missing_counts = df.select([count(when(isnull(col(c)) | isnan(col(c)), c)).alias(c) for c in numeric_cols])
    non_numeric_missing_counts = df.select([count(when(isnull(col(c)) | (trim(col(c)) == ""), c)).alias(c) for c in non_numeric_cols])

    missing_counts = numeric_missing_counts.join(non_numeric_missing_counts)
    return missing_counts

missing_counts = missing_values_count(df)
missing_counts.show()

# Convert missing_counts to Pandas DataFrame
missing_counts_pd = missing_counts.toPandas().transpose()
missing_counts_pd.columns = ["missing_count"]
missing_counts_pd["missing_percentage"] = (missing_counts_pd["missing_count"] / df.count()) * 100

print("Missing Data Percentage - mxmh:\n", missing_counts_pd)

# Visualize missing data heatmap
missing_data = df.toPandas().apply(lambda x: x.isnull() | x.isna() | (x.astype(str).str.strip() == ""))
plt.figure(figsize=(24, 12))
sns.heatmap(missing_data, cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap - mxmh')
plt.show()

```

```python
# Function to impute numerical columns with median
def impute_numerical(df, columns):
    for column in columns:
        median_value = df.approxQuantile(column, [0.5], 0.25)[0]
        if column in ["Anxiety", "Depression", "Insomnia", "OCD", "Age"]:
            median_value = int(median_value)  # Ensure whole numbers for these columns
        df = df.withColumn(column, when(isnull(col(column)) | isnan(col(column)), median_value).otherwise(col(column)))
    return df

# Function to impute categorical columns with mode
def impute_categorical(df, columns):
    for column in columns:
        mode_value = df.groupBy(column).count().orderBy('count', ascending=False).first()[0]
        df = df.withColumn(column, when(isnull(col(column)) | (trim(col(column)) == ""), mode_value).otherwise(col(column)))
    return df

# Columns to be imputed
numerical_columns = ["Age", "Hours per day", "BPM", "Anxiety", "Depression", "Insomnia", "OCD"]
categorical_columns = ["Primary streaming service", "While working", "Instrumentalist", "Composer", "Fav genre", "Exploratory", "Foreign languages", "Music effects",
                       "Frequency [Classical]", "Frequency [Country]", "Frequency [EDM]", "Frequency [Folk]", "Frequency [Gospel]", "Frequency [Hip hop]",
                       "Frequency [Jazz]", "Frequency [K pop]", "Frequency [Latin]", "Frequency [Lofi]", "Frequency [Metal]", "Frequency [Pop]", "Frequency [R&B]",
                       "Frequency [Rap]", "Frequency [Rock]", "Frequency [Video game music]", "Permissions"]

# Impute numerical and categorical columns
df = impute_numerical(df, numerical_columns)
df = impute_categorical(df, categorical_columns)

# Check for missing values after imputation
missing_counts_post_imputation = df.select([count(when(isnull(col(c)) | isnan(col(c)) | (trim(col(c)) == ""), c)).alias(c) for c in df.columns])
missing_counts_post_imputation_pd = missing_counts_post_imputation.toPandas().transpose()
missing_counts_post_imputation_pd.columns = ["missing_count"]

# Display the missing values for each column after imputation
missing_counts_post_imputation_pd["column"] = missing_counts_post_imputation_pd.index
missing_counts_post_imputation_pd.reset_index(drop=True, inplace=True)

print("Missing Values After Imputation:")
print(missing_counts_post_imputation_pd)

# Visualize the missing values after imputation
plt.figure(figsize=(15, 10))
sns.barplot(x='column', y='missing_count', data=missing_counts_post_imputation_pd)
plt.xticks(rotation=90)
plt.title('Missing Values Count After Imputation')
plt.xlabel('Columns')
plt.ylabel('Missing Count')
plt.show()

```
